```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from random import randrange
```

```python
bg_dir = "./BackgroundImage"
bg_name = f'flower{randrange(1,6)}.png'
bg_image = os.path.join(bg_dir, bg_name)
```

```python
print(os.path.isfile(bg_image))
```

```python
bg_image = cv2.imread(bg_image)

fig = plt.figure(num=1, figsize=(4,5))

rgb_bg = cv2.cvtColor(bg_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_bg)
plt.title(bg_name)
```

```python
img_color = cv2.resize(rgb_bg,(256,256))

plt.imshow(img_color)
plt.title(bg_name)
```

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    # initialize the dimensions of the image to be resized and
    # grab the image size
    dim = None
    (h, w) = image.shape[:2]

    # if both the width and height are None, then return the
    # original image
    if width is None and height is None:
        return image

    # check to see if the width is None
    if width is None:
        # calculate the ratio of the height and construct the
        # dimensions
        r = height / float(h)
        dim = (int(w * r), height)

    # otherwise, the height is None
    else:
        # calculate the ratio of the width and construct the
        # dimensions
        r = width / float(w)
        dim = (width, int(h * r))

    # resize the image
    resized = cv2.resize(image, dim, interpolation = inter)

    # return the resized image
    return resized
```

```python
resized_img = image_resize(rgb_bg, 256, 256)

plt.imshow(img_color)
plt.title(bg_name)
```
